```python
import torch
from transformers import MarianMTModel, MarianTokenizer
```

```python
print(f"CUDA Available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"GPU Device: {torch.cuda.get_device_name(0)}")
```

```python
MODEL_NAME = "Helsinki-NLP/opus-mt-en-ml"  # English to Malayalam Translation Model
MAX_LENGTH = 128
```

```python
model = MarianMTModel.from_pretrained(MODEL_NAME)
tokenizer = MarianTokenizer.from_pretrained(MODEL_NAME)
```

```python
model.eval()
```

```python
def translate_to_malayalam_colloquial(text):
    # Construct the translation prompt
    prompt = f"Translate to Malayalam: {text}"
    inputs = tokenizer(prompt, return_tensors="pt", max_length=MAX_LENGTH, truncation=True, padding=True)
    inputs = {k: v.to(model.device) for k, v in inputs.items()}

    # Generate translation
    outputs = model.generate(
        **inputs,
        max_length=MAX_LENGTH,
        do_sample=True,
        temperature=0.7,
        num_return_sequences=1,
        pad_token_id=tokenizer.pad_token_id,
        eos_token_id=tokenizer.eos_token_id,
        num_beams=1,
        top_p=0.95,
        repetition_penalty=1.2
    )

    # Decode and return the translation
    translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return translated_text
```

```python
test_sentences = [
    "What is your name?",
    "What is data structure?",
    "How does this work?",
    "Can you explain this to me?",
    "Where can I find the documentation?",
    "What is the error in this code?",
    "How do I fix this bug?",
    "Can you help me understand this?",
    "What is the best way to learn programming?",
    "Where can I practice coding?",
    "How do I install Python?",
    "What is the difference between AI and machine learning?",
    "Can you summarize this for me?",
    "Translate this sentence into Malayalam.",
    "What is the capital of India?",
    "How can I improve my problem-solving skills?",
    "What are the latest advancements in AI?",
    "Why is my code running slow?",
    "Can you generate a summary for this text?",
    "How do I use a for loop in Python?"
]

print("\nTesting the model:\n")
for sentence in test_sentences:
    translation = translate_to_malayalam_colloquial(sentence)
    print(f"English: {sentence}")
    print(f"Malayalam Colloquial: {translation}")
    print("-" * 50)
```

```python
from ipywidgets import interact, widgets
from IPython.display import display, HTML

def interactive_translation(text_input):
    if text_input:
        translation = translate_to_malayalam_colloquial(text_input)
        display(HTML(f"""
        <div style='padding: 10px; border-radius: 5px;'>
            <b>English:</b> {text_input}<br>
            <b>Malayalam Colloquial:</b> {translation}
        </div>
        """))
```

```python
interact(interactive_translation,
         text_input=widgets.Text(description='English:', placeholder='Enter text to translate'))
```

```python
import json
import pandas as pd

# Generate translations
dataset = []
for sentence in test_sentences:
    translation = translate_to_malayalam_colloquial(sentence)
    dataset.append({"English": sentence, "Malayalam_Colloquial": translation})

# Save as JSON file
with open("ml-colloquial-translator.json", "w", encoding="utf-8") as f:
    json.dump(dataset, f, ensure_ascii=False, indent=4)

# Save as CSV file (optional)
df = pd.DataFrame(dataset)
df.to_csv("ml-colloquial-translator.csv", index=False)
```
